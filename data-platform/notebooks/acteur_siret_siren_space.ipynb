```python
import sys
import os
from pathlib import Path

# On part du dossier du notebook: data-platform/notebooks
project_root = Path.cwd().parent   # => data-platform
sys.path.append(str(project_root))
sys.path.append(str(project_root / "dags"))

from dags.utils.django import django_setup_full
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

django_setup_full()
```

```python
from django.db import connections

# Fermer explicitement les connexions existantes
for conn in connections.all():
    conn.close()
```

```python
from qfdmo.models.acteur import Acteur, RevisionActeur

NBSP = "\u00A0"
SP = " "

for cls in [Acteur, RevisionActeur]:
    siret_1 = (cls.objects
        .filter(siret__in=[NBSP, SP])
    )
    print(f"{cls.__name__} siret_1.count(): {siret_1.count()}")
    siren_1 = (cls.objects
        .filter(siren__in=[NBSP, SP])
    )
    print(f"{cls.__name__} siren_1.count(): {siren_1.count()}")

```

```python
for cls in [Acteur, RevisionActeur]:
    cls.objects.filter(siret__in=[NBSP, SP]).update(siret='')
    cls.objects.filter(siren__in=[NBSP, SP]).update(siren='')
print("Done")
```
